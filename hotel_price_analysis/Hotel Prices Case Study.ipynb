# Case Study - Movement in Price as Check-in date Approaches 

A Product Owner at an online travel company was contemplating implementing 'urgency messages' on the website to encourage customers to complete their booking as fast as possible. Examples include:
* “Prices have been rising. Book now to lock in your rates!” 
* “Your check-in is fast approaching. Book now to lock in your rates!”

An analysis was requested  of the movement of the price as the day approaches the check-in date. The data was randomly sampled booking data from five different cities with check-in between 10/10/2016 – 12/31/2016.

### Deeper dive into the Top Performing Properties (Accomodation with the Most Bookings):

We choose a 'cut-off' for the number of bookings (ie a property has to have had a greater number of bookings than the cut-off for it to be included in the list)

```python
def top_hotels (data, cut_off):
    
    """
    The function plots a bar chart of the number of bookings per properties for properties that have bookings above a 
    threshold (cut-off point)
    
    """
        
    hotels_subset = hotels.query('count > @cut_off')
    hotels_subset['hotel_id'] = hotels_subset['hotel_id'].astype(str)
    plt.figure(figsize=(15,10))
    plt.bar(hotels_subset['hotel_id'],hotels_subset['count'], color='#00A9DC') 
    for i in range(1, len(hotels_subset['hotel_id'])+1):
        plt.text(i-1,hotels_subset['count'][i-1]/2,hotels_subset['count'][i-1], ha="center", fontsize=15)
    plt.title('No of Bookings for each Hotel (Top Performing Hotels)', fontsize=15, color="black")
    plt.xlabel('Hotel ID', fontsize=15)
    plt.ylabel('No of Bookings', fontsize=15)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=15)
    plt.show()
```

The function is run on the hotels dataset (of unique hotels) with a cut off of 500: 

```python
top_hotels (hotels, 500)
```

#### Fitting a Polynomial Regression Model with a degree of 4 (days before check-in vs price per night)

Trying to fit a better 'line of best fit' to the plots:

```python
for id in top_hotels['hotel_id']:
    
    data = df1[df1['hotel_id'] == id]

    x=data['days_before_checkin']
    y=data['ADR_USD']

    model = np.poly1d(np.polyfit(x, y, 4))
    polyline = np.linspace(-60, 0, 300)
                      
    plt.figure(figsize=(12,8))
    plt.scatter(x, y, color='#ff9999')
    plt.plot(polyline, model(polyline), color='#00A9DC', linestyle='--', linewidth=3) 
    plt.axhline(y = model(polyline).min(), color = 'black', linestyle = '-')
    plt.axhline(y = model(polyline).max(), color = 'black', linestyle = '-')
    
    plt.xlabel("Days Before Check in", fontsize=15)
    plt.ylabel("Average Daily Rate (USD)", fontsize=15)
    plt.title("Average Price Per Night (USD)", fontsize=15)
    plt.legend([id])
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.show()
    print(model)
    #plt.savefig(f"curve_plots_top_hotels for {id}.png")
```
